```python
import pandas as pd
import seaborn as sns
from sklearn.ensemble import IsolationForest
```

```python
df = pd.read_csv('learnt_new.csv')
sns.pairplot(df)
```

```python
X = df.values

clf = IsolationForest(contamination=0.03).fit(X)
clf.predict(X)
```

```python
df['out'] = clf.predict(X)
sns.pairplot(df, hue='out')
```
