```python
import lightkurve as lk
```

```python
kic = '4448777'
```

```python
lc = lk.search_lightcurve('KIC ' + kic, author="Kepler", quarter=[8,9,10], cadence="long").download_all().stitch()
```

```python
lc = lc.normalize()
lc.flux *= 1e6
lc.plot()
```

```python

pg = lc.to_periodogram(oversample_factor=1, normalization='PSD')
```

```python
pg.plot(scale='log');
```
